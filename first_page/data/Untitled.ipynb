```python
import pandas as pd
```

```python
# 重新读取数据，跳过第一行（年份标题行）
df = pd.read_csv('international-visitors-london.csv', skiprows=1)
```

```python
df
```

```python
# 删除全是 NaN 的列
df = df.dropna(axis=1, how='all')

# 删除最后一行（一般是总计）
df = df.iloc[:-1]
```

```python
df
```

```python
# 重命名第一列为 'Country'
df = df.rename(columns={'Country of origin': 'Country'})

# 转为长格式，便于可视化
df_long = df.melt(id_vars=['Country'], var_name='Year', value_name='Visitors')

# 转换数据类型
# 转换数据类型（先清除字符串中的逗号和非数字字符，再转 float）
df_long['Year'] = df_long['Year'].astype(str)

df_long['Visitors'] = (
    df_long['Visitors']
    .astype(str)
    .str.replace(",", "")  # 去掉千分位逗号（如 "2,233.10" -> "2233.10"）
    .str.replace(r"[^\d\.]", "", regex=True)  # 去掉非数字字符（如 "p"）
    .replace("", float("nan"))  # 空字符串处理为 NaN
    .astype(float)
)
```

```python
print(df_long[df_long['Country'] == 'USA'])
```

```python
df_long
```

```python
# 替换国家名映射字典
country_rename = {
    'Czech Republic': 'Czechia',
    'Hong Kong': 'Hong Kong S.A.R.',
    'Irish Republic': 'Ireland',
}

# 应用替换
df_long['Country'] = df_long['Country'].replace(country_rename)

# 删除 "Other" 开头的分类
df_long = df_long[~df_long['Country'].str.startswith("Other")]
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 读取 shapefile（更新为你的实际路径）
gdf = gpd.read_file("ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp")

# 画图看看世界地图
gdf.plot(figsize=(15, 10), edgecolor='black')

# 加标题
plt.title("World Countries - Natural Earth 110m", fontsize=16)
plt.show()
```

```python
# 计算每个国家的中心点经纬度
gdf['Longitude'] = gdf.geometry.centroid.x
gdf['Latitude'] = gdf.geometry.centroid.y

# 提取国家名称及经纬度
geo_df = gdf[['NAME', 'Latitude', 'Longitude']].rename(columns={'NAME': 'Country'})
```

```python
country_rename = {
    'United States of America':'USA',
}

# 应用替换
geo_df['Country'] = geo_df['Country'].replace(country_rename)
```

```python
geo_df
```

```python
# 获取两个表中的国家集合
countries_data = set(df_long['Country'].unique())
countries_geo = set(geo_df['Country'].unique())

# 找出数据中有但地理表中没有的国家
unmatched = sorted(countries_data - countries_geo)

# 显示不匹配的国家名（需人工检查命名差异，如缩写、大小写、简繁体等）
print("不匹配的国家名（数据中有，但地理信息中没有）：")
for c in unmatched:
    print(c)
```

```python
manual_coords = {
    'Bahrain': (26.0667, 50.5577),               # 中心点：首都麦纳麦附近
    'Hong Kong S.A.R.': (22.3193, 114.1694),     # 香港特别行政区
    'Singapore': (1.3521, 103.8198),             # 新加坡城市国家中心
}
```

```python
# 找出仍然缺失坐标的行
missing_coords = df_long[df_long['Country'].isin(manual_coords.keys())].copy()

# 为每一行填入经纬度
missing_coords['Latitude'] = missing_coords['Country'].map(lambda x: manual_coords[x][0])
missing_coords['Longitude'] = missing_coords['Country'].map(lambda x: manual_coords[x][1])

# 其余已匹配部分
existing_coords = pd.merge(df_long, geo_df, on='Country', how='inner')

# 合并所有有坐标的数据
df_geo = pd.concat([existing_coords, missing_coords], ignore_index=True)
```

```python
print(df_geo[df_geo['Country'] == 'USA'])
```

```python
df_geo = df_geo[['Country','Year','Visitors','Latitude','Longitude']]
```

```python
df_geo
```

```python
# 提取年份中的数字部分
df_geo['Year'] = df_geo['Year'].astype(str).str.extract(r'(\d{4})').astype(int)
```

```python
df_geo
```

```python
missing_coords = df_geo[df_geo[['Latitude', 'Longitude']].isnull().any(axis=1)]
print(missing_coords)
```

```python
df_geo.to_json("visitors.json", orient="records", lines=False, force_ascii=False, indent=2)
```

```python
# 假设 df_geo 已经准备好了并包含 Country, Latitude, Longitude 列

# 绘制世界底图
fig, ax = plt.subplots(figsize=(15, 10))
gdf.plot(ax=ax, edgecolor='black', facecolor='lightblue')

# 添加 df_geo 中的国家点
df_geo.plot.scatter(x='Longitude', y='Latitude', ax=ax, color='red', s=20)

# 标注国家名（可以加条件限制国家数目，避免重叠）
for _, row in df_geo.iterrows():
    ax.text(row['Longitude'] + 1, row['Latitude'] + 1, row['Country'], fontsize=8, color='darkred')

plt.title("Visitor Data Locations Overlaid on World Map", fontsize=16)
plt.show()
```

```python
# 生成柱状图使用的格式：[Visitors, Year, "", Country]
bar_data = []

for year in sorted(df_geo['Year'].unique()):
    subset = df_geo[df_geo['Year'] == year]
    top10 = (
        subset.sort_values(by='Visitors', ascending=False)
              .head(10)[['Visitors', 'Year', 'Country']]
              .copy()
    )
    top10['Blank'] = ''  # 第三列为空字符串
    bar_data.extend(top10[['Visitors', 'Year', 'Blank', 'Country']].values.tolist())
```

```python
# 导出为 JSON 文件（UTF-8）
import json
with open("bar_chart_data.json", "w", encoding="utf-8") as f:
    json.dump(bar_data, f, ensure_ascii=False, indent=2)
```
